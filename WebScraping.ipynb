```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://quotes.toscrape.com/'
response = requests.get(url)

print(response.status_code)  # Sprawdzenie statusu odpowiedzi
```

```python
print(response.text[:2000])  # Podgląd pierwszych 500 znaków zawartości strony
```

```python
print(response.headers)  # Nagłówki odpowiedzi
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
quotes = soup.find_all(class_='quote')
print(quotes)
```

```python
data = []
for quote in quotes:
    text = quote.find(class_='text')
    if text:
        data.append(text.get_text())

print(data)
```

```python
authors = soup.find_all(class_='author')
print(authors)
```

```python
data = []
for author in authors:
    text = author
    if text:
        data.append(text.get_text())
        print(text.get_text())
print(data)
```

```python
data = []
for i in range(1, 11):
  url = 'https://quotes.toscrape.com/page/' + str(i)
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  print(url)
  authors = soup.find_all(class_='author')
  for author in authors:
      text = author
      if text:
          data.append(text.get_text())
          print(text.get_text())

print(data)

```
